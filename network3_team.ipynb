```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y = pd.read_csv('datas/조이름.csv')
y
```

```python
print(y.keys)

teams = { i[0]: i[1:5] for i in y.to_numpy() }

print(teams)
```

```python
import networkx as nx

G = nx.Graph()
```

```python
for i in teams:
  print(teams[i])
  G1=nx.complete_graph(teams[i])
  G.add_edges_from(G1.edges)
```

```python
plt.figure(figsize=(10,10))
nx.draw_networkx(G, font_family='Malgun Gothic')
```
