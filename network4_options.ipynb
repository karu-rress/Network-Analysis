```python
import networkx as nx
import matplotlib.pyplot as plt


G1 = nx.Graph()
G2 = nx.Graph()
G3 = nx.Graph()
```

```python
G1.add_nodes_from(['고릴라','물개'])
G2.add_nodes_from(['고릴라','하마','펭귄','낙타'])

G1=nx.complete_graph(G1)
G2=nx.complete_graph(G2)
```

```python
nx.draw_networkx(G1, font_family='Malgun Gothic')
```

```python
nx.draw_networkx(G2, font_family='Malgun Gothic')
```

```python
G3.add_edges_from(G1.edges)
G3.add_edges_from(G2.edges)
nodecolor = ['red','red','blue','blue','blue']
G3.nodes
```

```python
nx.draw_networkx(G3, node_color=nodecolor, font_family='Malgun Gothic')
```

```python
nx.pagerank(G3).values()
```

```python
nx.draw_networkx(G3, node_color=list(nx.pagerank(G3).values()), font_family='Malgun Gothic')
```

```python
nx.draw_networkx(G3, node_color=list(nx.pagerank(G3).values()), cmap = 'jet_r', font_family='Malgun Gothic')
```

```python
nx.draw_networkx(G3, font_family='Malgun Gothic')
```

```python
G1.add_nodes_from([('고릴라', {'weight':'red'}), ('물개', {'weight':'black'})])
G2.add_nodes_from(['고릴라', '하마', '펭귄', '낙타'])

G1 = nx.complete_graph(G1)
G2 = nx.complete_graph(G2)
```

```python
G1.add_edges_from([('고릴라','펭귄',{'weight':5,'color':'green'})])
G1.add_edges_from([('고릴라','물개',{'weight':5,'color':'green'})])
```

```python
print(list(G1.edges(data='color'))[1])
```

```python
nx.draw_networkx(G1, edge_color=['red','green'], font_family='Malgun Gothic')
```

```python
pos = nx.shell_layout(G1)
```

```python
nx.draw_networkx(G1, pos=pos, font_family='Malgun Gothic')
```
