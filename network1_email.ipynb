```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#파일 불러오기
y = pd.read_table('datas/email-Enron.edges')
```

```python
print(len(y))
y
```

```python
#파일의 내용을 엣지리스트로 만들기
edgeslist = []

for i in y.to_numpy()[:10]:
  #print(i[0],i[1],i[2])
  edgeslist.append((i[0], i[1], {'weight' : i[2]}))
```

```python
#엣지리스트 확인하기
print(edgeslist)
```

```python
#엣지 가중치 리스트 만들기
edgesweight = []

for i in edgeslist :
  edgesweight.append(i[2]['weight'])

edgesweight
```

```python
print(edgeslist)
```

```python
import networkx as nx
#그래프 클래스 할당하기
G = nx.DiGraph()
```

```python
#엣지리스트 그래프에 추가하기
G.add_edges_from(edgeslist)
```

```python
#그래프 그리기
#링크 굵기는 가중치에  따라 다르게 한다.
plt.figure(figsize=(20,20))
nx.draw_networkx(G, width=edgesweight)
```

```python
W = nx.DiGraph()
W.add_weighted_edges_from(edgeslist)
```

```python
plt.figure(figsize=(20,20))
nx.draw_networkx(W, width=W)
```
